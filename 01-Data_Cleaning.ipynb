### Problem 1 - Skills Column: Case Duplicates & Fragmented Word Tokens

**Sub-problem A:** The scraper collects skills in the order WUZZUF renders them, which sometimes
includes the same skill twice: once from the "required" chip list and once from the
"preferred" list, with different casing.

Evidence:
```
Row skills: ['Information Technology (IT)', ... , 'Information technology (it)']
            ['Hardware Troubleshooting',    ... , 'Hardware troubleshooting']
```

**Sub-problem B:** The top "skills" by raw frequency are:  `|` (3979), `Development` (258),
`Software` (252), `Computer` (234) — these are NOT skills, they are word fragments
from multi-word skill names that were split incorrectly at some point in the pipeline.
The mean skill count per job is 25.7 with a max of 108, which is unrealistically high.

````python
SKILL_STOPWORDS = {'development', 'software', 'computer', 'science', 'it',
                   'information', 'technology', 'engineering', 'management'}

def filter_fragments(skills_str):
    skills =[s.strip().title() for s in str(skills_str).split(' | ') if s.strip()]
    skills =[s for s in skills if s.lower() not in SKILL_STOPWORDS and len(s) > 2]
    return [' , '.join(dict.fromkeys(skills))]

data['skills'] = data['skills'].apply(filter_fragments)
data['skills']
````

````python
# Preprocess step #1: Separate Technical Skills from Soft/Generic Skills

# Define the noise terms — generic labels that are categories, not skills
GENERIC_NOISE = {
    'Information Technology (IT)', 'Information Technology',
    'Computer Science', 'Software Development', 'Software Engineering',
    'Software', 'Development', 'Engineering', 'Technology', 'IT',
    'Technical', 'Management', 'Business', 'Science', 'Information',
    'Computer', 'Systems', 'Administration', '(IT)', 'Design',
}

# Define soft skills — valuable for separate soft-skills analysis
SOFT_SKILLS = {
    'Communication', 'Communication Skills', 'Problem Solving', 'Problem-solving',
    'Teamwork', 'Time Management', 'Time management', 'Presentation',
    'Leadership', 'Critical Thinking', 'Attention To Detail', 'Adaptability',
    'Customer Service', 'Customer Support', 'Sales', 'Negotiation',
    'Project Management',
}

def split_skills(skills_str):
    all_skills   = [s.strip() for s in str(skills_str).split(',') if s.strip()]
    technical    = [s for s in all_skills
                    if s not in GENERIC_NOISE and s not in SOFT_SKILLS
                    and len(s) > 2 and not s.isdigit()]
    soft         = [s for s in all_skills if s in SOFT_SKILLS]
    generic      = [s for s in all_skills if s in GENERIC_NOISE]
    return technical, soft, generic

# Apply and create three separate columns
results = data['skills'].apply(split_skills)
data['technical_skills'] = results.apply(lambda x: ', '.join(x[0]))
data['soft_skills']      = results.apply(lambda x: ', '.join(x[1]))
data['skill_count']      = results.apply(lambda x: len(x[0]))  # count of real tech skills
````

````python
# Preprocess step #2: Skill Alias Normalization

# These are all the same thing — without this, each counts separately
SKILL_ALIASES = {
    # Casing variants
    'Problem-solving':          'Problem Solving',
    'problem solving':          'Problem Solving',
    'Time management':          'Time Management',
    'Customer service':         'Customer Service',
    'Help desk support':        'Help Desk',
    'HelpDesk':                 'Help Desk',
    'Information technology (it)': 'Information Technology (IT)',
    # Tool name variants
    'React.js':                 'React',
    'React js':                 'React',
    'Reactjs':                  'React',
    'Node.js':                  'Node.js',        # keep .js convention
    'NodeJS':                   'Node.js',
    'JavaScript (ES6+)':        'JavaScript',
    'JS':                       'JavaScript',
    'Postgres':                 'PostgreSQL',
    'MS SQL':                   'SQL Server',
    'MSSQL':                    'SQL Server',
    'Cyber Security':           'Cybersecurity',
    'CyberSecurity':            'Cybersecurity',
    'Cyber security':           'Cybersecurity',
    'Quality Assurance':        'QA',
    'Quality Control':          'QC',
    'OOP':                      'Object-Oriented Programming',
    'REST':                     'RESTful APIs',
    'RESTful APIs':             'RESTful APIs',
    'Rest APIs':                'RESTful APIs',
    'ML':                       'Machine Learning',
    'AI':                       'Artificial Intelligence',
}

def normalize_skills(skills_str):
    skills = [s.strip() for s in str(skills_str).split(',') if s.strip()]
    normalized = [SKILL_ALIASES.get(s, s) for s in skills]
    return ', '.join(dict.fromkeys(normalized))  # deduplicate after normalization

data['technical_skills'] = data['technical_skills'].apply(normalize_skills)
````

````python
# Preprocess step #3: Skill Alias Normalization

data['skill_count'] = data['technical_skills'].apply(
    lambda s: len([x for x in s.split(',') if x.strip()]) if s else 0
)

data[['skills', 'technical_skills', 'soft_skills', 'skill_count']]
````

### Problem 3 — Salary Column: Mixed Data
The salary column has 4 distinct data types mixed in one field:
1. `'Confidential'` — 362 rows (79.2%)
2. `'Confidential, Benefits...'` — 51 rows (benefits/perks mixed in)
3. `'X to Y CUR Per Period'` — 44 rows (actual numeric salary)
4. Benefits text contains **typos**: `'Bouns'`, `'Boness'` (instead of "Bonus")

**New anomalies vs. previous dataset:**
- **HKD currency** (Hong Kong Dollar): `'15000 to 20000 HKD Per Month'` for a Cairo-based QA Associate at Mico World Limited — likely a foreign company paying in home currency. Needs special handling in normalization.
- **2 Per-Year salaries**: one in EGP (`363300 to 525000`), one in EUR (`19729 to 22054`). Must divide by 12 when normalizing to monthly.
- **1 Per-Hour salary**: `'700 to 1000 EGP Per Hour'` (AI Instructor). Must multiply by ~160 (working hours/month) to normalize.
- **2 flat/pinpoint salaries**: `min == max` (e.g., `10000 to 10000 EGP`) — these are single fixed values, not ranges.
- **Confidential, Confidential, ...** — one row has "Confidential" duplicated inside itself: `'Confidential, Confidential, Hybrid Work Mode...'`

````python
# Normalizing Confidential values to a common one
print("Confidential count before normalization: ", data[data['salary'] == "Confidential"].shape[0])
````

````python
data['salary'] = data['salary'].where(
    ~data['salary'].str.startswith('Confidential', na=False),
    other='Confidential'
)
````

````python
print("Confidential count after normalization: ", data[data['salary'] == "Confidential"].shape[0])
````

````python
# Parsing numeric values of salaries
sal_pat = r'([\d,]+)\s+to\s+([\d,]+)\s+(\w+)\s+Per\s+(\w+)'
parsed = data['salary'].str.extract(sal_pat)
parsed.columns = ['sal_min','sal_max','currency','pay_period']
data[['currency','pay_period']] = parsed[['currency','pay_period']]
data['sal_min'] = parsed['sal_min'].str.replace(',','').astype(float)
data['sal_max'] = parsed['sal_max'].str.replace(',','').astype(float)
data['salary_disclosed'] = data['salary'] != 'Confidential'
````

````python
# Exchange rates as of April 2026 — update as needed
EGP_RATES = {'EGP': 1.0, 'USD': 50.9, 'EUR': 55.2, 'HKD': 6.5, 'AED': 13.9}
PERIOD_MULTIPLIERS = {'Month': 1, 'Year': 1/12, 'Hour': 160}

data['sal_min_egp'] = (
    data['sal_min']
    * data['currency'].map(EGP_RATES).fillna(1)
    * data['pay_period'].map(PERIOD_MULTIPLIERS).fillna(1)
)
data['sal_max_egp'] = (
    data['sal_max']
    * data['currency'].map(EGP_RATES).fillna(1)
    * data['pay_period'].map(PERIOD_MULTIPLIERS).fillna(1)
)

data[data['salary_disclosed'] == True]
````

### Problem 4 — Posting Date Is Relative Text

The posting_date column contains relative text like `posted 2 hours ago`, `posted 2 days ago`, or `posted 1 month ago`. This is useless for time-series analysis.

````python
import re
from datetime import datetime, timedelta

SCRAPE_DATE = datetime(2026, 4, 19)

def parse_relative_date(s):
    m = re.search(r'(\d+)\s+(hour|day|week|month)', str(s))
    if not m: return pd.NaT
    n, unit = int(m.group(1)), m.group(2)
    deltas = {'hour': timedelta(hours=n), 'day': timedelta(days=n),
              'week': timedelta(weeks=n), 'month': timedelta(days=n*30)}
    return SCRAPE_DATE - deltas[unit]

data['posting_date_parsed'] = data['posting_date'].apply(parse_relative_date)
data['days_since_posted']   = (SCRAPE_DATE - data['posting_date_parsed']).dt.days

data[['posting_date', 'scraped_at', 'posting_date_parsed', 'days_since_posted']]
````

### Problem 9 - multivalues employment_type
Looking at the `employement_type` column carefully, this column is actually two completely different concepts packed together with `|`:

**Dimension 1 — Contract type:**
Full Time, Part Time, Freelance / Project, Shift Based, Volunteering

**Dimension 2 — Workplace mode:**
 On-site, Hybrid, Remote

````python
data['employment_type']
````

````python
# ── Step 1: Extract workplace mode (most analytically valuable) ──────
def get_workplace_mode(s):
    s = str(s)
    has_remote  = 'Remote'  in s
    has_hybrid  = 'Hybrid'  in s
    has_onsite  = 'On-site' in s

    if has_remote and not has_hybrid and not has_onsite:
        return 'Fully Remote'
    if has_hybrid:
        return 'Hybrid'
    if has_onsite:
        return 'On-site'
    return 'Unspecified'

data['workplace_mode'] = data['employment_type'].apply(get_workplace_mode)

# ── Step 2: Extract contract type (simplify to dominant) ─────────────
def get_contract_type(s):
    s = str(s)
    # Order matters: Full Time > Part Time > Freelance > others
    if 'Full Time' in s:
        return 'Full Time'
    if 'Part Time' in s:
        return 'Part Time'
    if 'Freelance / Project' in s:
        return 'Freelance'
    if 'Shift Based' in s:
        return 'Shift Based'
    if 'Volunteering' in s:
        return 'Volunteering'
    return 'Unspecified'

data['contract_type'] = data['employment_type'].apply(get_contract_type)

# ── Step 3: Boolean flags for precise multi-label rows ────────────────
# Use these when a single job explicitly advertises multiple options
data['is_full_time']  = data['employment_type'].str.contains('Full Time',        na=False)
data['is_part_time']  = data['employment_type'].str.contains('Part Time',        na=False)
data['is_freelance']  = data['employment_type'].str.contains('Freelance',        na=False)
data['is_remote']     = data['employment_type'].str.contains('Remote',           na=False)
data['is_hybrid']     = data['employment_type'].str.contains('Hybrid',           na=False)
data['is_onsite']     = data['employment_type'].str.contains('On-site',          na=False)

data[['employment_type', 'workplace_mode', 'is_full_time', 'is_part_time', 'is_freelance', 'is_remote', 'is_hybrid', 'is_onsite']]
````

### Problem 10 - `job_category` stacked values

Extract Primary + Secondary Signal
From the screenshot, the structure is always:

`[Primary Category] | [Secondary] | [Tertiary...]`

And 70%+ of rows start with IT/Software Development as the primary — meaning the primary category tells you almost nothing unique.

What is valuable is the secondary category, because it signals the domain/industry of the IT role:

- IT/Software Development | Sales/Retail       → IT Sales role
- IT/Software Development | Quality            → QA/Testing role
- IT/Software Development | Project/Program    → PM/Scrum role
- Installation/Maintenance | IT/Software Dev   → Field IT / Support role

````python
# ── Step 1: Normalize separator first ────────────────────────────────
data['job_categories'] = data['job_categories'].str.replace(
    r'\s*\|\s*', ' | ', regex=True
).str.strip()

# ── Step 2: Extract primary category (everything before first |) ──────
data['primary_category'] = data['job_categories'].str.split(' | ').str[0].str.strip()

# ── Step 3: Extract the non-IT secondary category as domain signal ────
IT_CATS = {'IT/Software Development', 'Engineering - Telecom/Technology',
           'Engineering - Other'}

def get_secondary_domain(cat_str):
    cats = [c.strip() for c in str(cat_str).split(' | ')]
    non_it = [c for c in cats if c not in IT_CATS]
    return non_it[0] if non_it else 'Pure IT'

data['job_domain'] = data['job_categories'].apply(get_secondary_domain)

# ── Step 4: Group secondary domains into meaningful buckets ───────────
DOMAIN_MAP = {
    'Pure IT':                      'Core IT',
    'Project/Program Management':   'IT Management',
    'Quality':                      'QA/Testing',
    'Creative/Design/Art':          'Design/UX',
    'Analyst/Research':             'Data/Analytics',
    'Installation/Maintenance/Repair': 'Field/Support IT',
    'Customer Service/Support':     'IT Support',
    'Sales/Retail':                 'IT Sales',
    'Business Development':         'IT Sales',
    'Accounting/Finance':           'IT Finance/ERP',
    'Marketing/PR/Advertising':     'IT Marketing',
    'Logistics/Supply Chain':       'IT Operations',
    'Purchasing/Procurement':       'IT Operations',
    'Operations/Management':        'IT Management',
    'Human Resources':              'HR Tech',
    'Training/Instructor':          'IT Training',
}
data['job_domain_group'] = data['job_domain'].map(DOMAIN_MAP).fillna('Other IT')

data[['job_categories', 'primary_category', 'job_domain', 'job_domain_group']]
````
